```python
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline

    from PIL import Image
    
    # Make sure that caffe is on the python path:
    caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
    import sys
    sys.path.insert(0, caffe_root + 'python')
    
    def showimage(im):
        plt.imshow(im)
    
    import caffe
    
    plt.rcParams['figure.figsize'] = (10, 10)
    plt.rcParams['image.interpolation'] = 'nearest'
    plt.rcParams['image.cmap'] = 'gray'
    
    net = caffe.Classifier(caffe_root + 'examples/mnist/mnist_kSAE_deploy.prototxt',
                           caffe_root + 'examples/mnist/mnist_kSAE.caffemodel')
    net.set_phase_test()
    net.set_mode_cpu()
```

```python
filters = net.params['decode1'][0].data;
fimage = np.zeros((1120,700));
a=0
b=0
for i in range(0,999):
    filt1 = filters[0][0,:,i];
    filt1 = filt1.reshape(28,28);
    filt1 -= filt1.min()
    filt1 /= filt1.max()
    fimage[a:a+28,b:b+28] = filt1;
    a = a+28;
    if a >= 1120:
        b = b+28;
        a = 0;
result = Image.fromarray((fimage * 255).astype(np.uint8))
showimage(fimage)
result.save('kSAE_decoder_filters.png')
```
